```python
import pandas as pd

#Lendo os arquivos

pedestres = pd.read_csv("PicMoney_Base_Pedestres.csv", sep=";")
transacoes = pd.read_csv("PicMoney_Base_de_Transacoes.csv", sep=";")
lojas = pd.read_csv("PicMoney_Lojas_e_Valores.csv", sep=";")
players = pd.read_csv("PicMoney-Base_Cadastral_de_Players.csv", sep=";")

for df in [pedestres, transacoes, lojas, players]: df.rename(columns=lambda x: x.strip(), inplace=True)
df['celular'] = df['celular'].str.strip()
lojas.rename(columns={'numero_celular': 'celular'}, inplace=True)
```

```python
capturas_por_usuario = lojas['celular'].value_counts().rename("n_capturas")
transacoes_por_usuario = transacoes['celular'].value_counts().rename("n_transacoes")
pedestres_por_usuario = pedestres['celular'].value_counts().rename("n_pedestre")
atividade = pd.concat([capturas_por_usuario,transacoes_por_usuario, pedestres_por_usuario], axis=1).fillna(0)
atividade = atividade.astype(int)
atividade['score_engajamento'] = atividade['n_capturas'] + 2 * atividade['n_transacoes'] + 0.5 * atividade['n_pedestre']

print(atividade.sort_values('score_engajamento', ascending=False).head(10))
```

```python
perfil_usuarios = atividade.merge(players, on='celular', how='left')

top_ativos = perfil_usuarios.sort_values('score_engajamento', ascending=False).head(20)
print(top_ativos[['celular', 'idade', 'sexo', 'bairro_residencial', 'categoria_frequentada', 'score_engajamento']])

print(top_ativos['idade'].describe())
print(top_ativos['sexo'].value_counts())

print(top_ativos['categoria_frequentada'].value_counts())
```

```python
import pandas as pd

transacoes = pd.read_csv("PicMoney_Base_de_Transacoes.csv", sep=";")
transacoes.columns = transacoes.columns.str.strip()
cupom_stats = transacoes.groupby('tipo_cupom').agg({'valor_cupom': ['mean', 'sum', 'count'], 'repasse_picmoney': ['mean', 'sum'] }).round(2)
cupom_stats.columns = ['valor_medio', 'valor_total', 'qtd_transacoes', 'repasse_medio_picmoney', 'repasse_total_picmoney']
cupom_stats = cupom_stats.sort_values('valor_total', ascending=False)

print(cupom_stats)
```

```python
categoria_stats = transacoes.groupby('categoria_estabelecimento').agg({ 'valor_cupom': 'mean', 'repasse_picmoney': 'mean', 'id_cupom': 'count' }).rename(columns={ 'valor_cupom': 'valor_medio', 'repasse_picmoney': 'repasse_medio', 'id_cupom': 'qtd_transacoes' }).sort_values('valor_medio', ascending=False)

print(categoria_stats.head(10))

campanha_stats = transacoes.groupby('id_campanha').agg({ 'valor_cupom': ['sum', 'mean'], 'repasse_picmoney': ['sum', 'mean'], 'id_cupom': 'count' }).round(2)

campanha_stats.columns = ['valor_total', 'valor_medio', 'repasse_total', 'repasse_medio', 'qtd_transacoes']
```

```python
print(campanha_stats.sort_values('valor_total', ascending=False).head(10))

transacoes['repasse_percentual'] = (transacoes['repasse_picmoney'] / transacoes['valor_cupom']) * 100

percentuais = transacoes.groupby('tipo_cupom')['repasse_percentual'].mean().round(2)
print(percentuais)
```

```python
import pandas as pd

pedestres = pd.read_csv("PicMoney_Base_Pedestres.csv", sep=";")
players = pd.read_csv("PicMoney-Base_Cadastral_de_Players.csv", sep=";")
lojas = pd.read_csv("PicMoney_Lojas_e_Valores.csv", sep=";")
transacoes = pd.read_csv("PicMoney_Base_de_Transacoes.csv", sep=";")

for df in [pedestres, players, lojas, transacoes]: df.columns = df.columns.str.strip()
df['celular'] = df['celular'].str.strip()
lojas.rename(columns={'numero_celular': 'celular'}, inplace=True)

locais_mais_frequentes = pedestres['local'].value_counts().head(10)
print("🏙️ Locais com mais pedestres:")
print(locais_mais_frequentes)
```

```python
transacoes_local = transacoes.merge(players[['celular', 'bairro_residencial']], on='celular', how='left')
fluxo = transacoes_local.groupby(['bairro_residencial', 'bairro_estabelecimento']).size().reset_index(name='qtd')
capturas = lojas.merge(players[['celular', 'bairro_residencial']], on='celular', how='left')
capturas_origem = capturas.groupby(['bairro_residencial', 'local_captura']).size().reset_index(name='qtd')
capturas_origem = capturas_origem.sort_values('qtd', ascending=False)
print(capturas_origem.head(10))
```

```python
pedestres_set = set(pedestres['celular'].unique())
compradores_set = set(transacoes['celular'].unique())

nao_convertidos = pedestres_set - compradores_set
print(f"🚶‍♂️ Usuários que circulam, mas não compram: {len(nao_convertidos)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fluxo_pivot = fluxo.pivot_table(index="bairro_residencial", columns="bairro_estabelecimento", values="qtd").fillna(0)
plt.figure(figsize=(12, 8))
sns.heatmap(fluxo_pivot, cmap="YlGnBu")
plt.title("Fluxo de Consumo: Moradia → Bairro de Compra")
plt.show()
```

```python
import pandas as pd

lojas = pd.read_csv("PicMoney_Lojas_e_Valores.csv", sep=";")
pedestres = pd.read_csv("PicMoney_Base_Pedestres.csv", sep=";")
transacoes = pd.read_csv("PicMoney_Base_de_Transacoes.csv", sep=";")
players = pd.read_csv("PicMoney-Base_Cadastral_de_Players.csv", sep=";")

lojas.rename(columns={'numero_celular': 'celular'}, inplace=True)
```

```python
for df in [lojas, pedestres, transacoes, players]:
    df.columns = df.columns.str.strip()
    for col in df.columns:
        if df[col].dtype == 'object':
            # Explicitly specify the date format
            if 'data' in col:
                df[col] = pd.to_datetime(df[col], format='%d/%m/%Y', errors='coerce')
            else:
                df[col] = df[col].astype(str).str.strip()


df_players_transacoes = pd.merge(transacoes, players, on='celular', how='left')
```

```python
df_players_trans_pedestres = pd.merge(df_players_transacoes, pedestres[['celular', 'data', 'horario', 'local']], on=['celular', 'data'], how='left', suffixes=('', '_pedestre'))

lojas.rename(columns={'data_captura': 'data'}, inplace=True)
```

```python
df_final = pd.merge(df_players_trans_pedestres, lojas[['celular', 'data', 'nome_loja', 'valor_cupom', 'tipo_loja']], left_on=['celular', 'data', 'nome_estabelecimento'], right_on=['celular', 'data', 'nome_loja'], how='left')

df_final.groupby('celular')['valor_cupom_y'].sum().sort_values(ascending=False).head(10)

top_users = df_final.groupby('celular')['valor_cupom_y'].sum().sort_values(ascending=False).head(100).index
df_final[df_final['celular'].isin(top_users)].groupby('idade')['valor_cupom_y'].sum().plot(kind='bar')
```
